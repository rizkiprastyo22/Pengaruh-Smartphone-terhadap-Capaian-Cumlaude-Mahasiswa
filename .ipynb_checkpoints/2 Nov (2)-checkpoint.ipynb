```python
df.drop(labels='aktifitas',axis=1,inplace=True)
```

#### Dilihat dari median responden yang memiliki nilai tengah 3.5, maka nilai tersebut dijadikan patokan untuk prediksi

```python
def impute_ipk(ipk):
    if ipk >= 3.5:
        return '>=3.5'
    else:
        return '<3.5'
df['ipk>=3.5']=df['ipk'].apply(impute_ipk)
```

```python
plt.title('Perbandingan Jumlah Responden >= dan < 3.5 Mahasiswa Semester 3 berdasarkan prodi')
sns.countplot(x='prodi',hue='ipk>=3.5',data=df[df['semester']==3])
```

# Data Manipulation

```python
df = pd.get_dummies(data=df,columns=['prodi'],drop_first=True)
```

```python
df = pd.get_dummies(data=df,columns=['meningkat'],drop_first=True)
```

```python
df = pd.get_dummies(data=df,columns=['semester'],drop_first=True)
```

# Data Train

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('ipk>=3.5',axis=1)
y = df['ipk>=3.5']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10)
```

# Data Modelling

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

# Report

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
